```python
def move_legend(ax, labels= False):
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    if labels:
        ax.legend(labels= labels, loc='center left', bbox_to_anchor=(1, 0.5), ncol= 3)
    else:
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol= 3)
        
def plot_line(df, ax, title= None, alpha= 1):
    df.plot(ax= ax, alpha= alpha)
    move_legend(ax)
    ax.set_title(title)

def plot(df, title= 'Confirmed', alpha= 1, edit= False):
    fig, axes = plt.subplots(3, 1, figsize= (8,3 * 3))
    for to_plot, title_, ax in [
                            (df, f'{title} Cumulative Frequency', axes[0]), 
                            (df.diff(), f'{title} Daily Frequency', axes[1]), 
                            (df.pct_change()*100, f'{title} Daily Frequency Growth', axes[2])
                          ]:
        plot_line(to_plot, ax, title_, alpha)
    plt.tight_layout()
    if edit:
        return fig, axes
    else:
        plt.show()
```

```python
gbConfirmed_100 = gbConfirmedSample.apply(lambda x: [np.nan if i < 100 else i for i in x])
gbConfirmed_100.iloc[40:50]
```

```python
def shift_na(ser):
    length= len(ser)
    name= ser.index
    ser= ser.dropna().tolist()
    while len(ser) != length:
        ser.append(np.nan)
    return pd.Series(ser, name)

gbConfirmedShifted = gbConfirmed_100.apply(shift_na)
gbConfirmedShifted.index = pd.Series(range(1, gbConfirmedShifted.shape[0] + 1), name= 'Days')
gbConfirmedShifted.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler

def scale(df, return_scaler= False):
    scaler = MinMaxScaler()
    if return_scaler:
        return scaler, pd.DataFrame(scaler.fit_transform(df), df.index, df.columns)
    else:
        return pd.DataFrame(scaler.fit_transform(df), df.index, df.columns)

gb11_scaled = scale(gb11)
gb11_scaled.style.bar()
```

```python
scaler_indonesia = MinMaxScaler().fit(gb11['Indonesia'].to_frame())

tes = pd.DataFrame(scale(gb30).mean(axis= 1), columns= ['mean'])
tes['Indonesia_scaled'] = gb11_scaled['Indonesia']
scaler, mean_scaled = scale(tes['mean'].iloc[:11].to_frame(), True)
tes['mean_scaled'] = mean_scaled
tes['prediction'] = scaler.transform(tes['mean'].to_frame())
tes['actual'] = gbConfirmedShifted['Indonesia']
tes['prediction_actual'] =  scaler_indonesia.inverse_transform(tes[['prediction']])
tes
```

```python
from sklearn.metrics import mean_squared_error, max_error

rmse = lambda true, pred: mean_squared_error(true, pred)**0.5

def predict(data):
    scaler = MinMaxScaler().fit(data.iloc[:11].to_frame())
    scaled_pred = scaler.transform(data.to_frame())
    pred = scaler_indonesia.inverse_transform(scaled_pred)
    return pred

def predict_mean_median(data): return predict(data.mean(axis= 1)).flatten(), predict(data.median(axis= 1)).flatten()

def score(pred_mean, pred_median):
    test = [893, 1046]
    train_rmse_mean= rmse(gb11['Indonesia'], pred_mean[:11])
    train_max_error_mean= max_error(gb11['Indonesia'], pred_mean[:11])
    train_rmse_median= rmse(gb11['Indonesia'], pred_median[:11])
    train_max_error_median= max_error(gb11['Indonesia'], pred_median[:11])
    test_rmse_mean= rmse(test, pred_mean[11:11+len(test)])
    test_max_error_mean= max_error(test, pred_mean[11:11+len(test)])
    test_rmse_median= rmse(test, pred_median[11:11+len(test)])
    test_max_error_median= max_error(test, pred_median[11:11+len(test)])
    return train_rmse_mean, train_max_error_mean, train_rmse_median, train_max_error_median, test_rmse_mean, test_max_error_mean, test_rmse_median, test_max_error_median

def plot_and_score(data, title= 0):
    actual= gbConfirmedShifted['Indonesia'].dropna().reset_index(drop= True)
    pred_mean, pred_median = predict_mean_median(data)
    fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize= (15, 3))
    ax1.plot(actual, c= 'r', label= 'Test')
    ax1.plot(gb11['Indonesia'].reset_index(drop= True), label= 'Train')
    ax1.plot(pred_mean, label= 'Mean of Samples', ls= '--', alpha= 0.5)
    ax1.plot(pred_median, label= 'Median of Samples', ls= '--', alpha= 0.5)
    ax1.set_title(f'Confirmed Prediction for {title} days')
    ax1.legend()
    ax1.set_xlabel('Days')
    ax1.set_ylabel('Confirmed Cases Cumulative Frequency')
    for ax, pred in zip([ax2, ax3], [pred_mean, pred_median]):
        sns.residplot(actual, pred[:len(actual)], lowess= True, ax= ax)
        ax.set_xlabel('Actual Value')
        ax.set_ylabel('Error')
    ax2.set_title('Mean Prediction Residual Plot')
    ax3.set_title('Median Prediction Residual Plot')
    plt.show()
    
    train_rmse_mean, train_max_error_mean, train_rmse_median, train_max_error_median, test_rmse_mean, test_max_error_mean, test_rmse_median, test_max_error_median = score(pred_mean, pred_median)
    
    print(f'''
Quanity of Samples = {data.shape[1]}
Mean of Samples
***************
train RMSE = {train_rmse_mean:.4f}
train max_error score = {train_max_error_mean:.4f}
test RMSE = {test_rmse_mean:.4f}
test max_error score = {test_max_error_mean:.4f}
prediction = {[('day {}'.format(i+11),j) for i, j in enumerate(pred_mean[11:].flatten())]}
Median of Samples
*****************
train RMSE = {train_rmse_median:.4f}
train max_error score = {train_max_error_median:.4f}
test RMSE = {test_rmse_median:.4f}
test max_error score = {test_max_error_median:.4f}
prediction = {[('day {}'.format(i+11),j) for i, j in enumerate(pred_median[11:].flatten())]}
--------------------
    ''')
```

```python
y_train = gbConfirmedShifted.loc[:10, 'Indonesia']
y_train
```

```python
y_test = gbConfirmedShifted.loc[11:, 'Indonesia'].dropna()
y_test
```

```python
Y= y_train.append(y_test)
Y
```

```python
from sklearn.preprocessing import PolynomialFeatures

x_train = y_train.index
x_test = y_test.index
print(x_test)

pn = PolynomialFeatures(degree= 5)
xp_train = pn.fit_transform(x_train.to_frame())
xp_test = pn.transform(x_test.to_frame())
xp_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

def print_score(true, pred):
    print(f'''
RMSE = {rmse(true, pred)}
max_error = {max_error(true, pred)}
    ''')
    
def pr_score(model):
    print(f'''Regression Coefficient = {model.named_steps['reg'].coef_}
*****************************************************
train
*****''', end= '')
    print_score(y_train, model.predict(x_train.to_frame()))
    print('''test
*****''', end= '')
    print_score(y_test, model.predict(x_test.to_frame()))

def pr(degree):    
    ii = degree
    pr = Pipeline([('poly', PolynomialFeatures(degree= ii)), ('reg', LinearRegression(fit_intercept= False))])
    pr.fit(x_train.to_frame(), y_train)    
    return pr

for degree in range(2, 8):
    print(f'.......degree of freedom = {degree}')
    model = pr(degree)
    pr_score(model)
```

```python
for ii in range(2, 8):
    model = pr(ii)
    fig, (ax1, ax2) = plt.subplots(1,2,figsize= (15,3.5))
    poly_result= pd.DataFrame({'prediction':model.predict(Y.index.to_frame()), 'Actual':Y.to_numpy()}, range(1, 14))
    poly_result.plot(alpha= 0.6, ax= ax1)
    sns.residplot(x= 'Actual', y= 'prediction', data= poly_result, lowess= True, ax= ax2)
    plt.show()
```

```python
pred_poly_a = pr(3).predict(Y.index.to_frame())
pred_poly_b = pr(4).predict(Y.index.to_frame())
pred_poly_avg = (pred_poly_a + pred_poly_b)/2
fig, (ax1, ax2) = plt.subplots(1,2,figsize= (15,3.5))
poly_result= pd.DataFrame({'prediction':pred_poly_avg, 'Actual':Y.to_numpy()}, Y.index)
poly_result.plot(alpha= 0.6, ax= ax1)
sns.residplot(x= 'Actual', y= 'prediction', data= poly_result, lowess= True, ax= ax2)
plt.show()
```

```python
print_score(Y, pred_poly_avg)
poly_result.apply(lambda x: x['prediction'] - x['Actual'], axis= 1)
```

```python
class pr_2_model_avg():
    def __init__(self, d1, d2, pr= pr):
        self.pr1 = pr(d1)
        self.pr2 = pr(d2)
        
    def predict(self, X):
        return (self.pr1.predict(X) + self.pr2.predict(X))/2

def pr_score_2_poly_model(d1, d2):
    model = pr_2_model_avg(d1, d2)
    print('''train
*****''', end= '')
    print_score(y_train, model.predict(x_train.to_frame()))
    print('''test
*****''', end= '')
    print_score(y_test, model.predict(x_test.to_frame()))
```
